```python
import pandas as pd
import numpy as np
```

```python
countries = ['China', 'Denmark', 'Austria', 'Korea, South', 'Czechia', 'United Kingdom', 'Australia', 'India', 'Brazil', 'Italy', 'Spain', 'Germany', 'France', 'US', 'South Africa']
countries2 = ['China', 'Denmark', 'Austria', 'South Korea', 'Czechia', 'United Kingdom', 'Australia', 'India', 'Brazil', 'Italy', 'Spain', 'Germany', 'France', 'United States', 'South Africa']
print(countries, len(countries))
```

```python
df_conf = pd.read_csv("./Data/time_series_covid19_confirmed_global.csv").drop(['Province/State', 'Lat', 'Long'], axis = 1)

df_death = pd.read_csv("./Data/time_series_covid19_deaths_global.csv").drop(['Province/State', 'Lat', 'Long'], axis = 1)

df_recov = pd.read_csv("./Data/time_series_covid19_recovered_global.csv").drop(['Province/State', 'Lat', 'Long'], axis = 1)
```

```python
df_L = []
```

```python
i = []

for index, row in df_conf.iterrows():
    if(row['Country/Region'] not in countries):
        i.append(index)
df_conf = df_conf.drop(i)

df_conf = df_conf.reset_index()


i = []
for index, row in df_death.iterrows():
    if(row['Country/Region'] not in countries):
        i.append(index)
df_death = df_death.drop(i)

df_death = df_death.reset_index()

i = []
for index, row in df_recov.iterrows():
    if(row['Country/Region'] not in countries):
        i.append(index)
df_recov = df_recov.drop(i)

df_recov = df_recov.reset_index()
```

```python
for c in countries:
    cList = [0, c]
    cList = cList + [0 for i in range(len(df_conf.columns) - 2)]
    for index, row in df_conf.iterrows():
        if(row['Country/Region'] == c):
            for i in range(2, len(df_conf.columns)):
                cList[i] = cList[i] + row[df_conf.columns[i]]
                
    runT = cList[2]
    for i in range(3, len(cList)):
        cList[i] = cList[i] - runT
        runT += cList[i]
        
    df = pd.DataFrame(columns = df_conf.columns[1:])
    df.loc[0] = cList[1:]
    
    df = df.melt(id_vars=["Country/Region"], var_name="Date", value_name = "Confirmed_Cases")
    
    df_L.append(df)
```

```python
ind = 0

for c in countries:
    cList = [0, c]
    cList = cList + [0 for i in range(len(df_recov.columns) - 2)]
    for index, row in df_recov.iterrows():
        if(row['Country/Region'] == c):
            for i in range(2, len(df_recov.columns)):
                cList[i] = cList[i] + row[df_recov.columns[i]]
                
    runT = cList[2]
    for i in range(3, len(cList)):
        cList[i] = cList[i] - runT
        runT += cList[i]
        
    df = pd.DataFrame(columns = df_recov.columns[1:])
    df.loc[0] = cList[1:]
    
    df = df.melt(id_vars=["Country/Region"], var_name="Date", value_name = "Recoveries")
    
    df_L[ind] = pd.merge(left = df_L[ind], right = df, how='inner')
    
    ind = ind + 1
```

```python
ind = 0

for c in countries:
    cList = [0, c]
    cList = cList + [0 for i in range(len(df_death.columns) - 2)]
    for index, row in df_death.iterrows():
        if(row['Country/Region'] == c):
            for i in range(2, len(df_death.columns)):
                cList[i] = cList[i] + row[df_death.columns[i]]
                
    runT = cList[2]
    for i in range(3, len(cList)):
        cList[i] = cList[i] - runT
        runT += cList[i]
        
    df = pd.DataFrame(columns = df_death.columns[1:])
    df.loc[0] = cList[1:]
    
    df = df.melt(id_vars=["Country/Region"], var_name="Date", value_name = "Deaths")
    
    df_L[ind] = pd.merge(left = df_L[ind], right = df, how='inner')
    
    ind = ind + 1
```
